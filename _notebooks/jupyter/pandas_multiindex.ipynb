# Pandas MultiIndex

The below examples are my summarized notes based on the [Pandas docs on advanced indexing](https://pandas.pydata.org/pandas-docs/stable/advanced.html). The documentation on Pandas is probably the best in the whole Python community. If you want to use the full power of Pandas yourself, [start reading](https://pandas.pydata.org/pandas-docs/stable)!

```python
import pandas as pd
import numpy as np
```

# Data

```python
categories = ['One', 'Two', 'Three', 'Four']
years = pd.date_range(end=pd.datetime.now(), periods=4, freq='A').year
```

```python
idx = pd.MultiIndex.from_product([categories, years], names=['category','year'])
```

```python
df = pd.DataFrame(data=np.random.rand(16,8),
                  index=idx,
                  columns=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])
```

```python
df
```

# MultiIndex Columns

I recomend creating you multi-header as a list of tuples, then using `pd.MultiIndex.from_tuples`, set it as the DataFrame columns.

```python
columns = [('Group1' if i < 4 else 'Group2',col) for i,col in enumerate(df.columns)]
```

```python
idx = pd.MultiIndex.from_tuples(columns, names=['group','subgroup'])
```

```python
df.columns = idx
```

```python
df
```

# Slicing

Pandas already has great documenation on [using slicers](https://pandas.pydata.org/pandas-docs/stable/advanced.html#using-slicers). Below if my simplified TLDR version.

**To allow slicing, you must first sort the index.**

> For MultiIndex-ed objects to be indexed & sliced effectively, they need to be sorted. As with any index, you can use `sort_index`. - [Pandas docs](https://pandas-docs.github.io/pandas-docs-travis/advanced.html#sorting-a-multiindex)

```python
df = df.sort_index()
```

**For the most convenient and logical slicing, use the builtin `pd.IndexSlice` tool**

```python
idx = pd.IndexSlice
```

**Example slice on all levels avialble.**

```python
df.loc[idx[['One', 'Two'],[2014,2015]],idx[['Group1', 'Group2'],['C','D','E']]]
```

**Slice on partial indices, and use `:` for accessing all others. Below is warning from the docs.**

> <mark>Warning</mark> You should specify all axes in the .loc specifier, meaning the indexer for the index and for the columns. There are some ambiguous cases where the passed indexer could be mis-interpreted as indexing both axes, rather than into say the MuliIndex for the rows. - [Pandas docs](https://pandas.pydata.org/pandas-docs/stable/advanced.html#using-slicers)

```python
df.loc[idx[:,[2016]],idx['Group2',:]]
```

# Reindexing

I recommend first creating a new index from your individually altered components, then reindex all levels at one time.

```python
new_categories = list(df.index.get_level_values('category').unique()) + ['Zero', 'Five']
new_categories
```

```python
new_years = list(range(df.index.get_level_values('year').min()-2, 
                       df.index.get_level_values('year').max()+2))
new_years
```

```python
idx = pd.MultiIndex.from_product([new_categories, new_years])
```

```python
df.reindex(idx, fill_value=0)
```
